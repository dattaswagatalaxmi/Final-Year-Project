```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Excel file
file_path = r"D:\JNWORK\yoursec_file.xlsx"  # Update with your file path
df = pd.read_excel(file_path)

# Assuming 'C%' column contains the C% values
c_percent_values = df['C%']

# Define C% value ranges
c_percent_ranges = range(0, 101, 10)

# Count sequences falling into each C% range
sequence_counts = []
for i in range(len(c_percent_ranges) - 1):
    count = ((c_percent_values >= c_percent_ranges[i]) & (c_percent_values < c_percent_ranges[i + 1])).sum()
    sequence_counts.append(count)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar([str(c) + '-' + str(c + 10) for c in c_percent_ranges[:-1]], sequence_counts, color='skyblue')
plt.xlabel('C% Value Ranges')
plt.ylabel('Number of Sequences')
plt.title('Number of Sequences vs. C% Value Ranges')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Read Excel file
file_path = r"D:\JNWORK\yoursec_file.xlsx"  # Update with your file path
df = pd.read_excel(file_path)

# Extract sequences from column 'A' and edit them
edited_sequences = []
for value in df['A']:
    if isinstance(value, str):  # Check if the value is already a string
        edited_sequence = value[2:-2]  # Remove first and last two letters
    else:
        edited_sequence = str(value)[2:-2]  # Convert to string and then manipulate
    edited_sequences.append(edited_sequence)

# Add new column 'Edited_seq' with edited sequences
df['Edited_seq'] = edited_sequences

# Save the modified DataFrame to a new Excel file or overwrite the existing file
output_file_path = r"D:\JNWORK\yoursec_file.xlsx" # Update with your desired output file path
df.to_excel(output_file_path, index=False)

print("New column 'Edited_seq' added to the DataFrame and saved to", output_file_path)
```

```python
# Read the Excel file
df = pd.read_excel(r"D:\JNWORK\Edited.xlsx")

# Define a function to calculate percentage of each element in a sequence
def calculate_percentage(sequence):
    sequence = str(sequence)  # Convert to string to handle non-string values
    c_count = sequence.count('C')
    h_count = sequence.count('H')
    e_count = sequence.count('E')
    total_count = len(sequence)
    
    # If any element is missing, set its percentage to 0
    if total_count == 0:
        return 0, 0, 0
    else:
        return (c_count / total_count) * 100, (h_count / total_count) * 100, (e_count / total_count) * 100

# Apply the function to each sequence in column 'B' and create new columns for percentages
df[['C%', 'H%', 'E%']] = df['B'].apply(lambda x: pd.Series(calculate_percentage(x)))

# Save the modified DataFrame back to the Excel file
df.to_excel('yourfir_file.xlsx', index=False)

```
