```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Excel file
file_path = r"D:\JNWORK\edited.xlsx"  # Update with your file path
df = pd.read_excel(file_path)

# Assuming 'C%' column contains the C% values
c_percent_values = df['C%']

# Define C% value ranges
c_percent_ranges = range(0, 101, 10)

# Count sequences falling into each C% range
sequence_counts = []
for i in range(len(c_percent_ranges) - 1):
    count = ((c_percent_values >= c_percent_ranges[i]) & (c_percent_values < c_percent_ranges[i + 1])).sum()
    sequence_counts.append(count)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar([str(c) + '-' + str(c + 10) for c in c_percent_ranges[:-1]], sequence_counts, color='skyblue')
plt.xlabel('C% Value Ranges')
plt.ylabel('Number of Sequences')
plt.title('Number of Sequences vs. C% Value Ranges')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
